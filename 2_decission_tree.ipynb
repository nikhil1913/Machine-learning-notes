```python
X = data.iloc[:,1:]
Y = data.iloc[:,0]
```

## preparing data for testing and training


```python

from sklearn.model_selection import train_test_split
X_train , X_test , Y_train, Y_test = train_test_split(X,Y,test_size= 0.2,random_state = 100)
```

## training our model here

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion= "entropy",random_state = 100,max_depth = 5,min_samples_leaf = 5)
model.fit(X_train,Y_train)
```

## we are testing our model here


```python

score = model.score(X_test,Y_test)
print(score)
```

## finding accuraccy and gettting classification report


```python
Y_pred = model.predict(X_test)
print(Y_pred)
```

```python
from sklearn.metrics import accuracy_score , classification_report

accuracy = accuracy_score(Y_test,Y_pred)
print(f"accuracy is : {accuracy*100} \n")
classification = classification_report(Y_test, Y_pred)
print(classification)
```
