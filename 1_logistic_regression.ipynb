```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn import metrics
digits = load_digits()
```

```python
plt.figure(figsize=(20,5))
for idx , (image,label) in enumerate(zip(digits.data[0:5] , digits.target[0:5])):
    plt.subplot(1,5, idx + 1)
    plt.imshow(np.reshape(image , (8,8)), cmap = plt.cm.gray)
    plt.title("train")
    plt.show()
```

```python
X_train ,X_test , Y_train ,Y_test = train_test_split(digits.data ,digits.target, test_size = 0.2 , random_state = 2)
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter = 500)

model.fit(X_train , Y_train)
```

```python
score = model.score(X_test,Y_test)
print(score)
```

```python
Y_pred = model.predict(X_test)
```

```python
cm = metrics.confusion_matrix(Y_test,Y_pred)
```

```python
print(cm)
```

```python
sns.heatmap(cm,annot = True, fmt=".0f",square = True, cmap= "Blues_r")
plt.ylabel("actual value")
plt.xlabel("predicted_value")
plt.show()
```
